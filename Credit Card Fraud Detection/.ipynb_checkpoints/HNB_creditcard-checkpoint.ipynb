```python
array = df.values
inf_indices = np.where(np.isinf(array))
nan_indices = np.where(np.isnan(array))
print(inf_indices, type(inf_indices))
print(nan_indices, type(nan_indices))
for row, col in zip(*inf_indices):
    array[row,col] = -1
    
for row, col in zip(*nan_indices):
    array[row,col] = 0
#array[]
X = array[:, 0:30]
y = array[:, 30]
ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X, y)

validation_size = 0.20
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res,
    test_size=validation_size, random_state=42)
```

# Hidden Naive Bayes

```python
hidden_nb_model = GaussianNB(var_smoothing=1e-7)
hidden_nb_model.fit(X_train, y_train)
```

```python
hidden_nb_predictions = hidden_nb_model.predict(X_test)
```

```python
report = classification_report(y_test, hidden_nb_predictions)
print("Performance Metrics:\n", report)
```
