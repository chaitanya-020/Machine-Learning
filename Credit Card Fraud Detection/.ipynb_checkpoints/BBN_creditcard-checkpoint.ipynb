```python
subset = df.sample(n=10000, random_state=42)
```

```python
subset
```

```python
X_subset = subset.drop('Class', axis=1)
y_subset = subset['Class']
```

```python
variable_mapping = {i: col for i, col in enumerate(X_subset.columns)}
renamed_X_subset = X_subset.rename(columns=variable_mapping)
```

```python
X_subset.rename(columns={'Class': 'Class_node'}, inplace=True)
```

```python
renamed_X_subset
```

```python
col=list(subset.columns)
col
```

```python
len(col)
```

```python
bbn_model = BayesianModel()
for i in range(len(col)-1):
    bbn_model.add_edge(col[i],col[i+1])
```

```python
bbn_model.nodes()
```

```python
bbn_model.fit(subset, estimator=BayesianEstimator)
```

```python
bbn_inference = VariableElimination(bbn_model)
```

```python
# elimination_order = [ 'V2', 'V3','V4','V5','V6','V7','V8','V9','V10'] 
```

```python
# query = bbn_inference.query(variables=['Class'], evidence=X_subset.iloc[0].to_dict())
```

```python
# query.values
```

```python
y_pred = []
for index, instance in X_subset.iterrows():
    query = bbn_inference.map_query(variables=['Class'], evidence=instance.to_dict())
    y_pred.append(query.values)
```
